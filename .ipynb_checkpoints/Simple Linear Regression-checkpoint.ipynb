#### We first import the required packages pandas, seaborn, numpy, sklearn, statsmodel to run the model

```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
```

#### Having imported our packages we then create a dataframe containing the content of the salary.csv .

```python
data = pd.read_csv('CSV/salary.csv')
```

```python
data.info()
```

####  After import our CSV we notice that one salary value is missing from the dataframe hence it has 513 records instead of the 514. We chose to drop the one row believing that one record could not be significant to cause a change as our dataframe is large enough that the one record isn't as great of information loss. An alternative method was to give the median amount to the one salary but that could change the outcome as we seek more information that is precise and not a median.

```python
data = data.dropna()
print(data.head())
```

#### We then print the head of the dataframe to see the content in the dataframe. 

#### We then draw up a null hypothesis that all the columns on here are not good predictors.

```python
data.describe()
```

#### We run a describe method on the dataframe to try and detect continuous data. From here we try to diagnose the continuous data 

```python
sns.pairplot(data)
```

#### From the pair plot we start to evaluate the data visualizations in scatter nd bar graphs to better see the relationship between each of the potential predictors. We can see that the Field, Male(gender column represented by 1 and 0s), position, other equal and degree are not continuous values and to continue to use such information for predicting is useless as the data will not be well predicted. Our null hypothesis as of right now is not still true as the rest of the columns are proving to be continuous data but that factor alone does not make a great predictor.

```python
data = data.drop(columns=['degree', 'otherqual', 'position', 'male', 'Field'])
print(data.head())
```

```python
sns.distplot(data['salary'],kde=False)
data.shape
```

```python
sns.boxplot(x=data["salary"])
```

#### We plot the distribution of the salary to see the range win which we have our salary and see that majority of the salary lies from around the 35 000 to the 60 000 range. The boxplot shows the 2 outliers that may skew. Not wanting to lose most of my data by dropping I decide to keep the 2 outliers and want to see the effect some predictors may have taken the outliers and test the relationship 

```python
for i in data.columns:
    sns.distplot(data[i],  kde=False,color='olive')
    plt.xlabel(f'{i}')
    plt.ylabel(f'Number of {i}')
    plt.title(f'Frequency of {i}')
    plt.show()
```

```python
data.corr()
```

```python
pearson_corr = stats.pearsonr(data['yearsworked'], data['salary'])
print(pearson_corr)
pearson_corr = stats.pearsonr(data['yearsrank'], data['salary'])
print(pearson_corr)
```

#### We then test the correlation each of the remaining columns have with salary as that is the dependent variable, what we are trying to predict with our model. We dismiss the salary 1 correlation as its matching the same values and move on to notice that yearsrank and yearsworked have strong correlation with salary with 0.623589 and 0.610706 for each respectively. From this, we can dismiss the rest and now prove our null hypothesis is, in fact, incorrect by saying none of our columns are strong indicators of being predictors. And accept our alternative hypothesis that one or more of the columns will be a great predictor with great correlation and having continuous data to predict from.

```python
sns.scatterplot(data['salary'], data['yearsworked'],color='olive')
plt.title('Salary and yearworked scatter plot points  ')
plt.show()
sns.scatterplot(data['salary'], data['yearsrank'],color='olive')
plt.title('Salary and yearsrank scatter plot points   ')
plt.show()
```

#### By checking distribution of data we decide to go with the Years worked as an indicator with it having more robust data to work with 

```python
x = data[['yearsworked']]
y = data['salary']
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.1,random_state = 10)
```

#### We then create a dataframe with Yearwworked as the only column as we have found our predictor which is the x the independent variable for the model. And for the y  variable the dependent we use salary as the series for its what we are trying to predict.

#### As we remember the regresstion model y = mx+c

#### We also split our data into a test set and training set. With the training set going to train our model to know about possible predictors using the 80/20 per cent split. The remaining 20 per cent is for shock to the model as its new information that will be predicted and see if our model can predict and test that yearsworked is thee predictor that has great significance in influencing salary increase or decrease 

```python
models = sm.OLS(y_train, x_train).fit()
print_model = models.summary()
print_model
```

```python
print('R2: ', models.rsquared *100,'%')
```

#### We have now trained our model using the training datasets(x_train and y_train). From this information, we notice that per increase in Years Worked the model predicts that an increase of 2856.16 per year as the coefficient. The R-squared being closer to 1 shows correlation that is statistically great but not great enough to call it thee predictor as with linear models a score of 80 and above will deem the model a good model. But it's good enough to try to predict. 
#### A high standard error shows that we may need a bigger dataset to train it as the standard error as of right now is quite, but if we increase the training set we end up overfitting and our model overtrained but if new information comes through the shock may cause the model to not operate well or predict well with new data. 
#### Our P-value being close to zero or in this case being 0 show that it is statistically significant significant 
#### The 0.25 and 0.95 percentile interval, with 25 per cent of the data believing that the increased minimum is 2696.49 and the largest amount at the 95% being 3015.83. This means that the value is within this range

```python
predictions = models.predict(x_test) 
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,models.predict(x_train),color='blue')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('TRAINING SET Predictions of values versus the actual test data set')
plt.show()

plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,models.predict(x_train),color='blue')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('TEST SET RESULT Predictions of values versus the actual test data set')
plt.show()
```

#### TRAINING SET - the red is the real values, straight from the training set, the blue is the predictions . its he best line after learning . from the first one we can see that most of the data is missed and the coefficient is way too far from the line of best fit from training .
#### TEST SET - the red is the test results this time from the x test no longer from the original data. This is what it hasn't learned as we can see the model is still weak as most of the point is far from the line of best fit created by the model. Though as few the line predicts well.


```python
print('TEST RMSE:', np.sqrt(sklearn.metrics.mean_squared_error(y_test, models.predict(x_test))))
print('TRAINING RMSE:', np.sqrt(sklearn.metrics.mean_squared_error(y_train, models.predict(x_train))))
```

#### Having looked at the RMSE for both the testing and training set, the gap between isn't significant but the fact that the training RMSE score is higher than the test means I have no overfit.

```python
predict_12_years = models.predict(12)
print(f'The predicted amount you will earn when working 12 years {predict_12_years}')
```

```python
predict_80_years = models.predict(80)
print(f'The predicted amount you will earn when working 80 years {predict_80_years}')
```
