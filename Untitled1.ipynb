#### We then test the corrolation each of the remaining columns have with salary as that is the dependent variable , what we are trying to predict with our model . #### By checking distribution of data we decide to go with the Years worked as an indicator with it having more robuist data to work with 

```python
x = data[['yearsworked']]
y = data['salary']
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.1,random_state = 10)
```

#### We then create a dataframe with Yearwworked as the only column as we have found our predictor which is the x the independent variable for model . And for the y the dependent we use salary as the series for its what we are trying to predict.

#### As we remember the regresstion model y = mx+c

#### We also split our data into a test set and training set . With the training set going to train our model to know about possible predictors using the 80/20 percent split. The remaining 20 perecent is for shock to the model as its new information that will be predicted and see if our model can predict and test that yearsworked is thee predictor that has great significance in influecing salary increase or decrease 

```python
models = sm.OLS(y_train, x_train).fit()
print_model = models.summary(alpha=0.05)
print_model
```

```python
print('R2: ', models.rsquared *100,'%')
```

#### We have now trained our model using the traing datasets(x_train and y_train) . From this information we notice that per increase in Years Worked the model predicts that an increase of 2856.16 per year as the coefficient. The R-squared being closer to 1 shows correlation that is statistically great but not great enough to call it thee predictor as with linear models a score of 80 and above will deem the model a good model . But its good enough to try to predict . 
#### A high standard error shows that we may need a bigger dataset to train it as the standard error as of right now is quite , but if we increase the training set we end up overfitting and our model overtrained but if new information comes through the shock may cause the model to not operate well or predict well with new data. 
#### Our P value being close to zero or in this case being 0 show that it is statistically significant significant 
#### The 0.25 and 0.95 percentile interval , with 25 percent of the data believing that the increase minimum is 2696.49 and the largest amount at the 95% being 3015.83.This means that the value is within this range

```python
predictions = models.predict(x_test) 
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,models.predict(x_train),color='blue')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('TRAINING SET Predictions of values versus the actual test data set')
plt.show()

plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,models.predict(x_train),color='blue')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('TEST SET RESULT Predictions of values versus the actual test data set')
plt.show()
```

#### TRAINING SET - the red are the real values , straight from the training set, the blue is the predictions . its he best line after learning . from the first one we can see that most of the data is missed and the coeff is way to far from the line of best fit from training .
#### TEST SET - the red are the test results this time from the x test no longer from the original data . This is what it hasnt learned as we can see the model is still weak as most of the point are far from the line of best fit created by the model. Though as few the line actually predicts well .


```python
print('TEST RMSE:', np.sqrt(sklearn.metrics.mean_squared_error(y_test, models.predict(x_test))))
print('TRAINING RMSE:', np.sqrt(sklearn.metrics.mean_squared_error(y_train, models.predict(x_train))))
```

#### Having looked at the RMSE for both the testing and training set , the gap between isnt significacnt but the fact that the training RMSE score is higher than the test means i have no overfit.
